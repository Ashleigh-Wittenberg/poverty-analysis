```python
directory = 'resources/'

# 2. List CSV files in the directory
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]

# 3. Loop through CSV files and read them into pandas DataFrames
dataframes = {}
for file in csv_files:
    file_path = os.path.join(directory, file)
    df = pd.read_csv(file_path)
    
    # 4. Store DataFrames in a dictionary
    dataframes[file] = df
```
